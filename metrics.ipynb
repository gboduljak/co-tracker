### Baseline

```python
import pandas as pd
import json
from glob import glob


models = ["baseline_online"]
flash_attn = ["true", "false"]
benchmarks = ["davis", "kinetics", "robotap", "stacking"]

dfs = []
for model in models:
    for attn in flash_attn:
        for benchmark in benchmarks:
            with open(
                f"./evals/{model}/flash_attention-{attn}/tapvid_{benchmark}_first/result_eval_.json"
            ) as fs:
                data = json.load(fs)
                data = {
                    key: [value]
                    for key, value in data.items()
                }
                data["model"] = [model]
                data["flash_attention"] = [attn]
                data["benchmark"] = [benchmark]
            dfs.append(pd.DataFrame(data))
```

```python
pd.concat(dfs)[["model", "flash_attention", "benchmark",
                "average_jaccard", "average_pts_within_thresh", "occlusion_accuracy"]]
```
